### 1. Write a program that takes a word as an input and print the number of vowels in the word.

```python
def print_vowels(word: str) -> int:
    """
    Take in a

    Parameters:
        word (str): input
    """
    VOWELS = ["A", "E", "I", "O", "U"]
    num_vowels = 0

    for char in word:
        if char.upper() in VOWELS:
            num_vowels += 1
        
    return num_vowels


print_vowels("runner")
```

### 3. Write a program that iterates from 1 to 20, printing each number and whether it's odd or even.

```python
for i in range(1, 21):
    if i % 2 == 0:
        print(f"{i} even")
    else:
        print(f"{i} odd")
```

### 4. Write a program to check if a string is a palindrome or not.

```python
def check_palindrome(word: str) -> bool:
    if len(word) == 0:
        return False

    l, r = 0, len(word) - 1

    while l < r:
        if (word[l].lower() == word[r].lower()):
            l += 1
            r -= 1
        else:
            return False
    
    return True


print(check_palindrome("civic"))
print(check_palindrome("word"))
```
